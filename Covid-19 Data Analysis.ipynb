```python
def return_country_wise_data(data, country_name):
    data_country = data[data["Country/Region"] == country_name].reset_index(drop = True)
    data_country.drop(["Province/State","Country/Region","Lat","Long"], axis = 1, inplace = True)
    data_country = data_country.T
    return data_country
```

```python
def covid_visualization(data, countries = []):
    plt.figure(figsize = (8,6))
    for country in countries:
        df_country = return_country_wise_data(data, country_name = country)
        df_country.index = pd.DatetimeIndex(df_country.index)
        num_of_cases = df_country.values
        dates = df_country.index.values
        plt.plot(dates, num_of_cases, label = country)
    plt.legend()
    plt.grid()
    plt.show()
    
        
```

```python
covid_visualization(data, countries = ['India','Afghanistan','Italy','Germany'])
```
